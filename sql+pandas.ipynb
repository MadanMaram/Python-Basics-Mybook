## Agg func

```python
""" select mean(Age), max(Age),min(Age)"""

customers.agg({'Age':['mean','max','min']})
```

# unique

```python
"""select DISTNICT Age from customers"""
customers.Age.unique()
#customers['Age'].unique
```

##  order by

```python
# order by
""" Select * from Customers where Genere = "Female" order by Spending_Score_0_100  """


customers[customers.genres == 'Comedy'].sort_values('Age')
```

## order by (Decending order)

```python
# order by (Decending order)
""" Select * from Customer where Genere = "Female" order by Spending_Score_0_100 DESC """


customers[customers.genres == 'Comedy'].sort_values('Age',ascending=False)
 
```

## Group BY(count)

```python
#Group BY(count)

"""select genres, Age, count(*) From customers GROUPBY genres, Age"""

#pandas equilant
customers.groupby(['genres','Age']).size().to_frame('Count').reset_index().sort_values('Age',ascending = False)
```

```python
customers.groupby(['genres','Age']).size()
```

```python
#Group BY(count and descending order on a column)

"""select genres, Age, count(*) From customers GROUPBY genres, Age"""

#pandas equilant
customers.groupby(['genres','Age']).size().to_frame('Count').reset_index()
```

## Having

```python
""" select age, count(*) from customers GROUPBY HAVING count(*) < 3"""

#pandas eqivalent
customers.groupby(['movieId']).filter(lambda g: len(g) < 3).groupby('movieId').size()
```

## IN

```python
"""select * from customers HERE Age in (20,30,40)"""

customers[customers.Age.isin([20,30,40])]
#customers.loc[(customers['Age'] == 20)|(customers['Age'] == 30)|(customers['Age'] == 40)]
```

```python
customers.Age.value_counts()
```

## Not In

```python
"""select * from customer WHERE Age NOT IN (10,20,30)"""

#pandas Eqivalent

customers[~customers.Age.isin([20,30,40])]
```

## Top N observations

```python
"""select * from customers OEDER BY Age DESC LIMIT 10"""

customers.nlargest(10, columns='Age')
```

## Top N observations with offset

```python
"""select * from customers OEDER BY Age DESC LIMIT 10 OFFSET 10"""

customers.nlargest(20, columns='Age').tail(10)
```

## Join

```python
"""select * from customers c oin shopcustomers t ON c.customerID = t.CustID where cgenres = "Female" """

#pandas eqivalent
customers.merge(shopcustomers[customers.genres == 'Drama'], on = 'Age',how = 'left')
```
